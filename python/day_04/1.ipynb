```python
class Apple:
    manufacturer='Apple Inc'
    contactWebsite='https://support.apple.com/en-in'

    def contactDetails(self):
        print('To contact us Visit: ',self.contactWebsite)
```

```python
class Macbook(Apple):
    def __init__(self):
        self.yearsOfManufacture=2020

    def manufactureDetails(self):
        print(f"This Macbook was manufactured in the year{self.yearsOfManufacture}")
        print(f"This Macbook was manufactured by {self.manufacturer}")

        
```

```python
macBook=Macbook()
macBook.contactDetails()
macBook.manufactureDetails()
```

```python

class Car:
    def __init__(self):
        self._color = "Black" 
        print(self._color) # Protected attribute by convention

class BMW(Car):
    def __init__(self):
        super().__init__()
        print("Protected attribute inside subclass:", self._color)

bmw = BMW()

# Technically possible (but discouraged) access from outside:
print("Discouraged external access to protected attribute:", bmw._color)
```

```python

class Car:
    def __init__(self):
        self.__yearOfManufacture = 2017  # Private (name-mangled)

    def getYearOfManufacture(self):
        return self.__yearOfManufacture

car = Car()

# Direct access raises an error
try:
    print(car.__yearOfManufacture)
except AttributeError as e:
    print("Direct access error:", e)



print(car.getYearOfManufacture())
# Accessing using name mangling (not recommended)
print("Access via name mangling:", car._Car__yearOfManufacture)
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    """Abstract base class for all shapes."""

    @abstractmethod
    def area(self):
        """Return area of the shape."""
        pass

class Square(Shape):
    """Concrete implementation of Shape for squares."""

    def __init__(self, side):
        self.side = side

    def area(self):
        # super().area()
        return self.side * self.side
    
class Rectangle(Shape):
    """Concrete implementation of Shape for rectangle"""
    
    def __init__(self,length,width):
        self.length=length
        self.width=width

    def area(self):
        return self.length*self.width
square=Square(5)
print(square.area())
rectangle=Rectangle(10,2)
print(rectangle.area())
```

```python
from abc import ABC,abstractmethod

class Employee(ABC):
    def login(self):
        print("Employee logged in")

    @abstractmethod
    def calculate_salary(self):
        pass

class FullTimeEmployee(Employee):

    def calculate_salary(self):
        return 50000

fte=FullTimeEmployee()
fte.login()
print("salary:",fte.calculate_salary())   


# intern=Employee()
# print("salary:",fte.calculate_salary())   

```
